It‚Äôs like converting something from one form to another ‚Äî for example, turning a number into text or text into a number.

**Real-Life Example**

Imagine you wrote your age as "20" on paper (text).
If someone wants to do math with it, they can‚Äôt ‚Äî they must convert it to a number (20) first.

Python automatically decides the data type, but sometimes you need to change it yourself ‚Äî for example:

When reading input using input() (it gives strings)

When doing math with text numbers like "5"

When formatting or combining text and numbers

# What Is an if‚Äìelse Statement?

**In simple words:**
if‚Äìelse lets your program make decisions ‚Äî just like humans do.

It allows your code to choose between two or more actions depending on whether a condition is True or False.

**Real-Life Example**

**Imagine:**

If it‚Äôs raining, take an umbrella.
Otherwise, wear sunglasses.

That‚Äôs an if‚Äìelse decision!

````python
raining = True
if raining:
    print("Take an umbrella")
else:
    print("Wear sunglasses")
````

````python
raining =False

if raining:
    print("Take an umbrella")
else:
    print("Wear sunglasses")
````

**Basic Syntax:**

```Python
if condition:
    # code runs if condition is True
else:
    # code runs if condition is False


````python
# Example 1: Simple if Statement
age = 18

if age >= 18:
    print("You are an adult.")
````

````python
# Example 2: if‚Äìelse
age = 15

if age >= 18:
    print("You can vote.")
else:
    print("You are too young to vote.")
````

**Example 3: if‚Äìelif‚Äìelse (Multiple Conditions)**

If you want to check more than one condition, use elif (which means else if).

````python
marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: F")
````

````python
# Example 4: Nested if

# You can put an if inside another if.

age = 20
has_id = True

if age >= 18:
    if has_id:
        print("You can enter.")
    else:
        print("You need an ID.")
else:
    print("You are too young.")
````

````python
# Example 5:Using if with Logical Operators

# You can combine conditions using and, or, not.

age = 25
citizen = True

if age >= 18 and citizen:
    print("You can vote.")
else:
    print("You cannot vote.")
````

````python
# Example 6: Short if‚Äìelse (Ternary Operator)

# You can write an if‚Äìelse in one line (for simple cases).
age = 18
status = "Adult" if age >= 18 else "Minor"
print(status)
````

````python
age = 17
status = "Adult" if age >= 18 else "Minor"
print(status)
````

````python
number = int(input("Enter a number: "))

if number > 0:
    print("Positive")
elif number == 0:
    print("Zero")
else:
    print("Negative")
````

## 2. while Loop
### When to use:

Use a while loop when you want to repeat something until a condition becomes false.

Layman example:

````python
water = 0

while water < 10:
    print("Filling...")
    water += 1
````

A **set** in Python is like a **bag that automatically removes duplicates**.

- You can put many items inside.
- Python will **keep only unique items**.
- The order of items does **not** matter.
- You cannot access items using indexing (like list[0]).

### Example:
```python
numbers = {1, 2, 3, 3, 2}
print(numbers)  
# Output: {1, 2, 3}


**Why Sets Are Useful?**

<li>To remove duplicates from data

<li>To check if something exists (fast lookup)

<li>To compare groups of items

<li>To perform math-like operations (union, intersection, difference)

## üîó Set Operations (VERY Important)

6. *`union()`*

Combines two sets (no duplicates).



````python
a = {1, 2}
b = {2, 3}
a.union(b)  # {1, 2, 3}
````

Keeps only common items.



````python
a.intersection(b)  # {2}
````

Items that are in A or B but not in both.



````python
a.symmetric_difference(b)  # {1, 3}
````
